**IMPORTING MODULES**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

**LOAD DATASETS**

```python
data_sets = pd.read_csv ("Salary_dataset.csv")
```

**ASSIGNING THE FEATURES AND TARGET TO A VARIABLES**

```python
X = data_sets[["YearsExperience"]]
Y = data_sets['Salary']
```

**INITIALIZE AND TRAIN THE MODEL**

```python
model = LinearRegression()
model.fit(X,Y)
```

**PREDICT THE MODEL**

```python
pred = model.predict(X)
```

**SAVE THE MODEL**

```python
import joblib
joblib.dump(model , "Salary_pred_model.pkl")
```

**PLOTTING THE GRAPH**

```python
import matplotlib.pyplot as plt

# Plot the actual data points from your dataset
plt.scatter(X, Y, color='red', label='Actual Data')

# Plot the professional regression line (the model's predictions)
plt.plot(X, model.predict(X), color='blue', linewidth=2, label='Regression Line')

# Add titles and labels so it looks professional
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

# Display the graph
plt.show()
```
